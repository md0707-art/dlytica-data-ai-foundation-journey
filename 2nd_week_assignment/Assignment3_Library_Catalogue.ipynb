```python
# Assignment 3 - Library Book Catalogue

class Book:

    # Class Attribute
    total_books = 0

    # Constructor
    def __init__(self, title, author, genre, available=True):
        self.title = title
        self.author = author
        self.genre = genre
        self.available = available

        Book.total_books += 1

    # Class Method: Create Object from Dictionary
    @classmethod
    def from_dict(cls, data):
        return cls(
            data["title"],
            data["author"],
            data["genre"]
        )

    # Class Method: Total Books Count
    @classmethod
    def get_total(cls):
        return f"Total books registered: {cls.total_books}"

    # Borrow Book
    def borrow(self):
        if not self.available:
            raise ValueError(f"'{self.title}' is already borrowed.")
        self.available = False

    # Return Book
    def return_book(self):
        if self.available:
            raise ValueError(f"'{self.title}' is not currently borrowed.")
        self.available = True

    # String Representation
    def __str__(self):
        status = "✓" if self.available else "✗"
        return f"[{status}] {self.title} | {self.author} | {self.genre}"


# Main Program
try:

    # Create Books
    b1 = Book("Palpasa Cafe", "Narayan Wagle", "Novel")

    b2 = Book("Seto Dharti", "Amar Neupane", "Novel")

    b3 = Book.from_dict({
        "title": "Summer Love",
        "author": "Subin Bhattarai",
        "genre": "Romance"
    })

    b4 = Book("Karnali Blues", "Buddhisagar", "Novel")

    # Borrow Two Books
    b1.borrow()
    b2.borrow()

    print("Library Catalogue")
    print("-" * 60)

    print(b1)
    print(b2)
    print(b3)
    print(b4)

    print("\nReturning a Book...")
    b1.return_book()

    print(b1)

    print("\n" + Book.get_total())

    # Test Error Handling
    print("\nTesting Exception Handling:")
    b2.borrow()

except ValueError as e:
    print("Error:", e)
```
