```python
#Assignment 1 — Student Grade Analyser Python
# Task 1: Letter Grade Function
def letter_grade(score):
    if score < 0 or score > 100:
        raise ValueError("Score must be between 0 and 100.")
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Task 2: Analyse Student Data Function
def analyse(students):
    analyzed_data = {}
    for name, score in students.items():
        analyzed_data[name] = {
            "score": score,
            "grade": letter_grade(score)
        }
    return analyzed_data

# Task 3: Summary Function
def summary(results):
    # Print Individual Student Results
    for name, data in results.items():
        print(f"{name:<7}: {data['score']} → {data['grade']}")
    
    print() # Formatting spacer
    
    # Calculate Metrics
    scores = [data["score"] for data in results.values()]
    class_avg = sum(scores) / len(scores)
    
    highest_scorer = max(results, key=lambda k: results[k]["score"])
    lowest_scorer = min(results, key=lambda k: results[k]["score"])
    
    # Calculate Grade Counts
    grade_counts = {}
    for data in results.values():
        grade = data["grade"]
        grade_counts[grade] = grade_counts.get(grade, 0) + 1
        
    # Print Aggregated Summary Metrics using f-strings
    print(f"Class Average : {class_avg:.2f}")
    print(f"Highest Score : {highest_scorer} ({results[highest_scorer]['score']})")
    print(f"Lowest Score  : {lowest_scorer} ({results[lowest_scorer]['score']})")
    
    # Formatting counts display exactly to match constraints
    counts_str = f"A={grade_counts.get('A', 0)}  B={grade_counts.get('B', 0)}  C={grade_counts.get('C', 0)}  D={grade_counts.get('D', 0)}  F={grade_counts.get('F', 0)}"
    print(f"Grade Counts  : {counts_str}")

# Task 4: Main Execution and Exception Handling Wrapper
try:
    # Testing Valid Sample Data
    students = {
        "Alice": 92, "Bob": 78, "Carol": 85,
        "Dave": 61, "Eve": 55, "Frank": 99
    }
    results = analyse(students)
    summary(results)
    
    print("\n--- Testing Invalid Score Trigger ---")
    # This call should fail and jump directly to the except block safely
    letter_grade(110)

except ValueError as e:
    print(f"❌ Handled ValueError: {e}")
```
