```python
# Assignment 2 - Personal Expense Tracker

# Task 1: Load Expenses from File
def load_expenses(filename):
    expenses = []

    try:
        with open(filename, "r") as file:
            for line in file:
                category, amount = line.strip().split(",")
                expenses.append((category, float(amount)))

    except FileNotFoundError:
        print("File not found. Starting with an empty expense list.")
        return []

    return expenses


# Task 2: Add Expense
def add_expense(filename, category, amount):
    if amount <= 0:
        raise ValueError("Amount must be greater than 0.")

    with open(filename, "a") as file:
        file.write(f"{category},{amount}\n")


# Task 3: Category Totals
def category_totals(expenses):
    totals = {}

    for category, amount in expenses:
        totals[category] = totals.get(category, 0) + amount

    print("\nCategory Total")
    print("-" * 30)

    grand_total = 0

    for category, total in totals.items():
        print(f"{category:<15}: RS{total:.2f}")
        grand_total += total

    print("-" * 30)
    print(f"Grand Total     : RS{grand_total:.2f}")


# Task 4: Expenses Above Threshold
def above_threshold(expenses, limit):
    return [(category, amount) for category, amount in expenses if amount > limit]


# Main Program
filename = "expenses.txt"

# Create sample data file
with open(filename, "w") as file:
    file.write("Food,150\n")
    file.write("Transport,60\n")
    file.write("Food,200\n")
    file.write("Entertainment,120\n")
    file.write("Utilities,300\n")
    file.write("Transport,45\n")

try:
    # Load expenses
    expenses = load_expenses(filename)

    # Display totals
    category_totals(expenses)

    # Show expenses above threshold
    limit = 100
    print(f"\nExpenses above RS{limit}:")

    expensive_items = above_threshold(expenses, limit)

    for category, amount in expensive_items:
        print(f"{category:<15} → RS{amount:.2f}")

    # Test ValueError
    print("\nTesting invalid expense:")
    add_expense(filename, "Books", -50)

except ValueError as e:
    print("Error:", e)
```
